```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread('/content/drive/MyDrive/prawesti.jpg')

#mengetahui ukuran panjang dan lebar dari gambar
plt.imshow(img)
```

```python
# OpenCV membaca image dan menyimpan dalam channel warna BGR 
# (Blue Green Red) 
img2 = img
img3 = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img3)
```

```python
# menampilkan citra Grayscale
img_gray = cv.imread('/content/drive/MyDrive/prawesti.jpg', cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)

```

```python
# menampilkan colormap dengan warna 'gray'
plt.imshow(img_gray, cmap='gray')
```

```python
# menampilka citra grayscale dengan colormap warna 'magma'
plt.imshow(img_gray, cmap='magma')
```

```python
# menampilkan RGB di resize dalam ukuran panjang 512 dan lebar 1024
img4 = cv.resize(cv.cvtColor(img,cv.COLOR_BGR2RGB), (512,1024))
plt.imshow(img4)
```

```python
# citra RGB ditampilkan dalam ukuran lebih besar posisi terbaik
img5 = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB),0)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.imshow(img5)
```

```python
# membuat geometri 2D
#membuat black image denga tipe data int16
black_image= np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_image)
```

```python
# menambahkan bentuk persegi panjang sesuai koordinat pt1 dan pt2
cv.rectangle(black_image, pt1=(384,0), pt2=(510,150), color=(0, 255,0), thickness=10)
plt.imshow(black_image)
```

```python
# menambahkan bentuk persegi panjang sesuai koordinat pt1 dan pt2
cv.rectangle(black_image, pt1=(200,200), pt2=(300,300), color=(0, 0,255), thickness=15)
plt.imshow(black_image)
```

```python
# menambahkan bentuklingkaran sesuai radius
cv.circle(black_image, center=(100,100), radius= 50, color=(255, 0,0), thickness=8)
plt.imshow(black_image)
```

```python
# menambahkan garis sesuai koordinat pt1 dan pt2
cv.line(black_image, pt1=(0,0), pt2=(512,512), color=(255, 255,255), thickness=5)
plt.imshow(black_image)
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_image, text='Prawesti', org=(10,500),fontFace=font, fontScale=4, color=(255, 255, 0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black_image)
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_image, text='Prawesti', org=(10,500),fontFace=font, fontScale=4, color=(255, 255, 0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black_image)
```

```python
# membuat geometri 2D
#membuat black image denga tipe data int32
black_image2= np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_image2)
```

```python
vertices= np.array([[100, 300], [200, 200], [400, 300], [200, 400]], dtype=np.int32)
vertices
```

```python
pts = vertices.reshape((-1, 1, 2))
pts
```

```python
cv.polylines(black_image2,[pts], isClosed = True, color=(255, 0, 0), thickness = 5)
plt.imshow(black_image2)
```
